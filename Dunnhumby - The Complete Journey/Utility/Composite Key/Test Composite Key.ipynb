```python
import Connection as C
import key_finder as k
import pandas as pd
import json
```

```python
file = open("details.txt", "r")
content = file.readlines()
file.close()
schema_name = content[0].split('=')[1].strip()
q = f'''
SELECT 
    tablename
FROM 
    pg_catalog.pg_tables 
WHERE 
    schemaname='{schema_name}'
;
'''
c = C.Connection()
c.start()
c.run(q)
rows = c.cursor.fetchall()
c.stop()
d = {}
for table in rows:
    if table[0] not in d:
        d[table[0]] = ''
    K = k.Key('public', table[0])
    K.run()
    d[table[0]] = ' + '.join(K.comp_key)
```

```python
d
```

```python
with open(f'comp_key_{schema_name}.json', 'w') as fp:
    json.dump(d, fp)
```
